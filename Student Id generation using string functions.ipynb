"""
## **Case Study – Student ID Card Generator**

**Instructions:**

• Create variables: student_name, roll_number, department  

• Take input from the user using input() function  

• Convert student_name and department into uppercase using upper()  

• Extract first 4 letters of student_name and department using slicing  

• Create Student ID by combining name, department, and roll number  

• Display Name, Department, Roll Number, and Student ID only  

• Use print() and f-string for output display    
"""

```python
#Student Id card Generator

#variables creation using input()
student_name=input("Enter the student name: ").upper()
roll_number=input("Enter the roll number: ")
department=input("Enter the department: ").upper()

# ID number generation
name=student_name[:4]
dept=department[:4]
student_id=name+dept+roll_number

# print("\n -------display the student details------")
print(f"Name: {student_name}")
print(f"Department: {department}")
print(f"Roll Number: {roll_number}")
print(f"Student ID: {student_id}")
```
