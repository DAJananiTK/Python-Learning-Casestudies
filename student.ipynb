**Instructions:**

1. Create variables:

   student_name

2. Use input() function to get student details from the user.

```python
# Student Registration system

#creating variables using input function an typecasting
student_name=input("Enter the student name: ")
student_age=int(input("Enter the student age: "))
course_name=input("Enter the course name: ")
course_fee=float(input("Enter the course fee: "))
print("\n Display student details")

#gst calculation
GST=course_fee*0.18
Total_fee=course_fee+GST

# display student details
print(f"Student Name : {student_name}")
print(f"Student Age : {student_age}")
print(f"Course Name : {course_name}")
print(f"Course Fee : {course_fee}")
print(f"GST_calculated : {GST}")
print(f"Total Fee : {Total_fee}")
print("\n Datatypes")

#display datatype of variables
print(type(student_name))
print(type(student_age))
print(type(course_fee))
```
