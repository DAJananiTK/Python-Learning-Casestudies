# **COMPLILER  VS INTERPRETER**


| Feature        | Compiler                                          | Interpreter                             |
| -------------- | ------------------------------------------------- | --------------------------------------- |
| Meaning        | Converts entire program into machine code at once | Converts and executes code line by line |
| Speed          | Faster execution                                  | Slower execution                        |
| Error Checking | Shows all errors after compilation                | Stops when an error occurs              |
| Output         | Creates executable file                           | Does not create executable file         |
| Execution      | Program runs after compilation                    | Runs directly                           |
| Examples       | C, C++                                            | Python, JavaScript                      |

This is the VARIABLE / IDENTIFIER / PARAMETER


It acts like a label or container name pointing to some data in memory.

You can use it later in the program to access the value.

**Right Side**

This is the VALUE / LITERAL / ARGUMENT

Python stores this exact data in memory and links it to the variable Name.

# **Type Casting**

**What is Type Casting?**

Type casting means converting one datatype into another datatype.

**int() → Convert to Integer**

```python
x = 30.5

y =int(x)
print(y)
print(type(y))
```

**float() → Convert to Decimal**

```python
x = 30
y =float(x)

print(y)
print(type(y))
```

**str() → Convert to String**

```python
x = 45
y =str(x)

print(y)
print(type(y))
```

**bool() → Convert to Boolean**

```python
x = 0
print(bool(x))
```

# **Formatted String Literal (f-string) in Python**

**What is Formatted String Literal?**

Formatted String Literal (f-string) is used to insert variables directly inside a string.

```python
#Without f-string
name = "Ravi"
age = 25

print("My Name is :",name,"\nMy age is :",age)


```

```python
#Basic f-string
print(f"My Name is {name} My age is {age}")
```

```python
#Decimal Formatting
temp = 45.5656

print(f"Temperature is :{temp:.2f}")
```
