```python
#Library fine calculator

#  creating variables using input() and typecasting
student_name=input("Enter the student name: ")
book_name=input("Enter the book name: ")
days_late=int(input("Enter the count of days late: "))
fine_per_day=float(input("Enter the fine per day value: "))
print("\n Display details")

#total fine calculation
Total_fine=days_late*fine_per_day

#displaying the details
print(f"Student Name: {student_name}")
print(f"Book Name: {book_name}")
print(f"Number of days late: {days_late}")
print(f"Fine per day: {fine_per_day}")
print(f"Total Fine: {Total_fine}")

#display the  datatypes
print("\n Datatypes")
print(type(student_name))
print(type(book_name))
print(type(days_late))
print(type(fine_per_day))
```
