```python
#Email Username Generator


#variables creation using input()
full_name=input("Enter your full name: ").lower().replace(" ",".")
domain_name=input("Enter your domain name: ")
email=full_name + "@" + domain_name

print("\n------------ Display name and email---------")
print(f"Full name: {full_name.title().replace("."," ")}")
print(f"Generated email: {email}")







```
