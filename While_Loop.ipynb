## **Case Study – Password Verification System**

**Instructions:**

1. Create a variable named `password` with a default password value.

2. Take password input from the user using input().

3. Use a while loop to:
   • Check whether the entered password matches the original password  
   • Display "Incorrect password... try again" if password is wrong  
   • Ask user to re-enter password until correct password is entered  

4. Display "Access granted" when the correct password is entered

```python
password="jan0204@"
user_input=input("Enter the password : ")
while password != user_input:
  print("Incorrect password.........Try again")
  user_input=input("Enter the password : ")
else:
  print("Access granted")

```

## **Case Study – ATM PIN Verification System**

**Instructions:**

1. Create a variable named `correct_pin` with a default PIN number.

2. Use a while loop to repeatedly ask the user to enter the PIN.

3. Check whether the entered PIN matches the correct PIN.

4. If the PIN is correct:
   • Display "Login Successful"  
   • Stop the loop using break  

5. Otherwise:
   • Display "Incorrect PIN... Try Again"

```python
correct_pin=7744

while True:
    user_pin = int(input("Enter your PIN: "))

    if user_pin == correct_pin:
        print("Login Successful")
        break
    else:
        print("Incorrect PIN... Try Again")

```

## **Case Study – Password Verification System with Limited Attempts**

**Instructions:**

1. Create variables:
   • password  
   • attempts  
   • max_attempts  

2. Use a while loop to allow the user to enter the password.

3. Check whether the entered password matches the original password.

4. If the password is correct:
   • Display "Access granted!"  
   • Stop the loop using break  

5. If the password is incorrect:
   • Increase attempt count  
   • Display remaining attempts  

6. If maximum attempts are reached:
   • Display "Maximum attempts reached. Access denied!"

```python
password="mithra123"
attempts=0
max_attempts=3
while attempts<max_attempts:
  user_pswd=input("Enter the password: ")
  if user_pswd == password:
    print("Access granted")
    break
  else:
    attempts+=1
    remaining=max_attempts-attempts
    print(f"Incorrect password.attempts remaining:{remaining}")
if attempts==max_attempts:
  print("Maximum attempts reached. Access denied!")
```

## **Case Study – Student Marks Filter System using continue**

**Instructions:**

1. Create two lists:
   • students  
   • marks  

2. Use a while loop to iterate through the student list.

3. Check student marks using if condition.

4. Use continue statement to:
   • Skip students who scored 50 or above  

5. Display only students who scored below 50.

6. Print student name and marks using f-string

```python
students=["rana","surya","nila","megha","aarav"]
marks=[78,36,74,45,89]
i=0

while i<len(students):
  if marks[i]>=50:
    i+=1
    continue
  else:
    print(f"{students[i]} scored {marks[i]} marks")
  i+=1
```
