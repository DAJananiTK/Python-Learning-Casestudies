## **Case Study – Student Course Registration System using Set**

**Instructions:**

1. Create a set named `courses` with course names:
   Python, SQL, Power BI, Python, Excel

2. Perform set operations:
   • Add course using add()  
   • Add multiple courses using update()  
   • Remove course using remove() and discard()

3. Display all courses and total number of courses using len()

```python
#Student Course Registration System using Set
 #set creation
courses={"Python","SQL","Power BI","Python","Excel"}
print(f"Courses: {courses}")

#set operations
# adding course
courses.add("Software Testing")
courses.add("Cybersecurity")
print(f"Courses: {courses}")

#adding multiple courses using update
courses.update(["Python","AI"])
print(f"Courses: {courses}")

#removing course using remove()
courses.remove("Excel")
print(f"Courses: {courses}")

# remove using discard()
courses.discard("AI")
print(f"Courses: {courses}")
print("\n ----------------------")


#display all courses
print(f"Courses after discarding: {courses}")
print(f"Total number of courses: {len(courses)}")




```
