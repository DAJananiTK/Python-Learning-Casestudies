# **Case Study – Student Details using Tuple (Packing & Unpacking)**

**Instructions:**

1. Create a tuple named `student` and store the following details:
   name, age, course

2. Unpack the tuple into separate variables:
   name, age, course

3. Display each value separately using print().

```python
#student details using tuple
#packing
students=("janani",28,"Data Analyst")
#unpacking
name,age,course=students
#displays the each value separately
print("---------Student Details ---------")
print(f"Name: {name}")
print(f"Age: {age}")
print(f"Course: {course}")
print(students)
print(type(students))

```

# **Case Study – Online Order Summary using Tuple**

**Instructions:**

1. Create a tuple named `order` with the following details:

2. Unpack the tuple into separate variables:
   order_id, product, quantity, price, status
