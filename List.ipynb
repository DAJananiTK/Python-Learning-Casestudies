# **Case Study – Shopping Cart System**

**Instructions:**

1. Create a list named `cart` with initial items:
   Milk, Bread, Eggs, Apple

2. Perform list operations:

   • Add multiple items using extend()

   • Insert item at index 2 using insert()

   • Remove item Apple using remove()

   • Remove last item using pop()

3. Use list methods:

   • index() to find position of Eggs

   • count() to count occurrences of Milk

4. Apply indexing:

   • Display first item

   • Display last item

5. Apply slicing:

   • Display first 3 items from the list

6. Find total number of items using len()

7. Display all outputs using print()

```python
#Shopping cart system

#initial list
cart=["Milk","Bread","Eggs","Apple"]
print(cart)

#list Methods
#extend
cart.extend(["Orange","Curd","Paneer","cheese","butter","banana"])
print(cart)
#insert
cart.insert(2,"Sauce")
print(cart)
#remove
cart.remove("Apple")
print(cart)
#pop
cart.pop()
print(cart)

#index()
print(cart.index("Eggs"))
# count
print(cart.count("Milk"))

# display first item
print(cart[0])

#display last item
print(cart[-1])

#slicing (first three items)
print(cart[:3])

#length
length=len(cart)
[print(f"Length={length}")]
#display the final output
print(f"Cart: {cart}")
```

# **Case Study – Student Marks Analysis**
**Instructions:**

Create a list named marks by getting 5 subject marks from the user using map() and list()

Perform calculations:

• Find total marks using sum()

• Find average marks using

• Calculate over all percentage assuming each subject is out of 100

```python
students_marks = list(map(int,input("Enter the students marks ").split(",")))
print(f"students_marks:{students_marks}")
sum_marks=sum(students_marks)
print(f"Total marks={sum_marks}")
avg=sum_marks/len(students_marks)
print(f"Average={avg}")
overall_total=len(students_marks)*100
percentage=(sum_marks/overall_total)*100
print(f"Percentage={percentage}%")
```
