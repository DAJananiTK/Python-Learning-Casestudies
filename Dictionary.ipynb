## **Case Study – Student Data Management System**

**Instructions:**

1. Create a dictionary named `student` with the following keys:
   name, age, course

2. Store multiple students' data using lists as values.

3. Add one new student details to the dictionary:

   • Append new name to name list  

   • Append new age to age list  

   • Append new course to course list  

4. Display all student details using dictionary keys.

5. Access and display:

   • First student's details using indexing

   • Newly added student's details using indexing

```python
#dictionary creation using list to store multiple values

student={
    "name":["rekha","rahini","janani","kalpana"],
     "age":[30,22,29,28],
    "course":["Data Analytics","Python","Cybersecurity","software testing"]
}


#add students details to the dictionary using append
student["name"].append("gayathri")
student["age"].append(24)
student["course"].append("SQL")

#display all student details using keys
print(f"Names: {student['name']}")
print(f"Age: {student['age']}")
print(f"Courses: {student['course']}")
print("\n -----------------------")

#display first student details
print(f"Name: {student['name'][0]}")
print(f"Age: {student['age'][0]}")
print(f"Course: {student['course'][0]}")
print("\n -----------------------")

#display newly added student details
print(f"Name: {student['name'][-1]}")
print(f"Age: {student['age'][-1]}")
print(f"Course: {student['course'][-1]}")



```
